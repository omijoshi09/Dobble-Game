# DobbleDeck class

- This class uses emoji file as external source and it uses random package
- This class has 10 methods:

    #### A. dobblegenerator()
- It will generate a logic to create a deck that the number pass to it in this case its value is 8
- This will also pass the value to DobbleCard class
#### B. emoji_generator()
- It will create dictionary of emoji
#### C. check_deck_auth()
- This method will check if the deck is valid or not
- If the verbose its true it will print every value by comparing to it.
#### D. add_card()
- This will return random card
#### E. remove_card()
- It will remove the card
#### F. validate_cards()
- It will check if the value is present within it
#### G. ask_user()
- This will take the number of iteration which deck can repeat
#### H. start_game()
- It will contrls the game operation and stors the result of both player
#### I. play_card()
- This will print the respective emoji and check if they are valid one
#### J. get_winner_input()
- return the value of the winner

```python
from random import choice
import emoji

class DobbleDeck:
    def __init__(self):
        self.deck_cards_is_valid = True
        self.dobblegenerator()
        # If Verbose is True
        #self.check_deck_auth(self.deck_cards, True)
         # If Verbose is False
        self.check_deck_auth(self.deck_cards)
        if self.deck_cards_is_valid:
            self.used_cards = []
            self.played_card = []
            self.emoji_generator()
            self.ask_user()
        else:
            print("Sorry, Deck is invalid!!")
    
    # it will generatethe logic for cards
    def dobblegenerator(self):
        self.deck_cards = []
        nIm = 8
        n = nIm - 1
        r = range(n)
        rp1 = range(n+1)
        c = 0

        # First card
        card_value_inner = set()
        c += 1
        for i in rp1:
            card_value_inner.add(i + 1)
        self.deck_cards.append(DobbleCard(c, card_value_inner))
        
        #n following cards
        for j in r:
            c = c+1
            card_value_inner = set()
            card_value_inner.add(1)
            for k in r:
                card_value_inner.add(n+2 + n*j +k)    
            self.deck_cards.append(DobbleCard(c, card_value_inner))

        # n x n following cards
        for i in r:
            for j in r:
                c = c+1
                card_value_inner = set()
                card_value_inner.add(i + 2)
                for k in r:
                    card_value_inner.add((n+1 +n*k + (i*k+j) % n)+1)
                self.deck_cards.append(DobbleCard(c, card_value_inner))
                
    # create image dictionary             
    def emoji_generator(self):
        self.imageDict = dict()
        fin = open('emoji_names.txt',"r")
        lines = fin.readlines()
        for i, el in enumerate(lines):
            self.imageDict[i+1] = emoji.emojize(el.strip())
    
    def check_deck_auth(self, deck, verbose = False):
        for outer_card_value, outer_card_details in enumerate(deck):
            outer_card_value = outer_card_value +  1
            outer_emoji_card = outer_card_details.card[str(outer_card_value)]
            if verbose:
                for card_no_inner in range(outer_card_value, len(deck)):
                    card_emojis_inner = deck[card_no_inner].card[str(card_no_inner + 1)]
                    print('Comparing the value', outer_emoji_card, 'with', card_emojis_inner)
                    if len(card_emojis_inner.intersection(outer_emoji_card)) != 1:
                        print('Sorry Invalid pair found')
                        self.deck_cards_is_valid = False
                        break
                    else:
                        print('Pair is valid')
            else:
                for card_no_inner in range(outer_card_value, len(deck)):
                    card_emojis_inner = deck[card_no_inner].card[str(card_no_inner + 1)]
                    if len(card_emojis_inner.intersection(outer_emoji_card)) != 1:
                        self.deck_cards_is_valid = False
                        break
            if not self.deck_cards_is_valid:           
                break

    
    def add_card(self):
        return choice(self.deck_cards)
    
    def remove_card(self):
        if self.played_card:
            self.played_card.pop()
    
 
    def is_valid(self, card1_no, card2_no, card2):
        if card1_no in self.used_cards or card2_no in self.used_cards:
            return False
        elif card1_no == card2_no:
            return False
        else:
            self.used_cards.append(card1_no)
            self.remove_card()
            self.played_card.append(card2)
            return True
    
           
            
    def ask_user(self):
        user_input = input("How many cards (Should be less than 56)? ")
        while True:
            if user_input.isdigit() and (int(user_input) > 1 or int(user_input) <= 56):
                break
            else:
                print('Please enter a valid input.')
            user_input = input("How many cards (<56)? ")    
        print("If you want to record a draw type 'd' or 'D'.")
        print()
        self.start_game(user_input)
        
    def start_game(self, user_input):
        player_a_score, player_b_score = 0, 0
        total_rounds = int(user_input)
        for rounds in range(1, total_rounds + 1):
            self.play_game()
            winner = self.get_winner_input()
            if winner == 'a':
                player_a_score += 1
            elif winner == 'b':
                player_b_score += 1
            print()    
        print("Score")
        print("A:", player_a_score)
        print("B:", player_b_score)
        
        if player_a_score > player_b_score:
            print('Player A wins')
        elif player_a_score == player_b_score:
            print('Match Tied !!')
        else:
            print('Player B wins')
            
    def play_game(self):
        while True:
            if not self.used_cards:
                card1 = self.add_card()
                card2 = self.add_card()
            else:
                card1 = self.played_card[0]
                card2 = self.add_card()
            card1_no = list(card1.card.keys())[0]
            card2_no = list(card2.card.keys())[0]
            
            if self.is_valid(card1_no, card2_no, card2):
                break
                
        card1_emoji_list = list(card1.card[card1_no])
        card2_emoji_list = list(card2.card[card2_no])
        for count in range(3):
            if count == 0:
                print(self.imageDict[int(card1_emoji_list[0])], end="")
                print(self.imageDict[int(card1_emoji_list[1])], end="")
                print(self.imageDict[int(card1_emoji_list[2])], end="")
                print("\t\t", end="")
                print(self.imageDict[int(card2_emoji_list[0])], end="")
                print(self.imageDict[int(card2_emoji_list[1])], end="")
                print(self.imageDict[int(card2_emoji_list[2])], end="")
                print()
            elif count == 1:
                print(self.imageDict[int(card1_emoji_list[3])], end="")
                print(self.imageDict[int(card1_emoji_list[4])], end="")
                print(self.imageDict[int(card1_emoji_list[5])], end="")
                print("\t\t", end="")
                print(self.imageDict[int(card2_emoji_list[3])], end="")
                print(self.imageDict[int(card2_emoji_list[4])], end="")
                print(self.imageDict[int(card2_emoji_list[5])], end="")
                print()
            else:
                print(self.imageDict[int(card1_emoji_list[6])], end="")
                print(self.imageDict[int(card1_emoji_list[7])], end="")
                print("\t\t", end="")
                print(self.imageDict[int(card2_emoji_list[6])], end="")
                print(self.imageDict[int(card2_emoji_list[7])], end="")
                print()       
        
        
    
    def get_winner_input(self):
        winner = input("Who wins (A or B)? ").strip().lower()
        while True:
            if winner.isdigit() or winner not in 'abd' or len(winner) != 1:
                 print('Please enter a valid input.')
            else:
                return winner
            winner = input("Who wins (A or B)? ").strip().lower()
```
